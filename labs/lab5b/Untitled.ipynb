```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = []
y = []
with open('prod.dat') as f:
    data = f.readlines()

for el in data:
    el = el.strip()
    x.append(int(el.split()[0]))
    y.append(float(el.split()[1]))
```

```python
plt.clf()
plt.plot(x,y)
plt.show()
```

```python
fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit) 
```

```python
plt.plot(x,y, '-', x, fit_fn(x), '-')
plt.show()
```

```python
plt.clf()
plt.hist(y, bins=100)
plt.show()
```
