```python
import numpy as np
from matplotlib import pyplot as plt
with open("./LJoutput.dat") as f:
    data1 = f.read()
    
with open("./LJoutput2.dat") as f:
    data2 = f.read()

data1 = data1.split('\n')
data2 = data2.split('\n')
data1 = [row.replace(" ", "")[1:-1].split(',') for row in data1][:-1]
data1 = np.array([[float(el) for el in row] for row in data1])

data2 = [row.split(' ') for row in data2][:-1]
data2 = np.array([[float(el) for el in row] for row in data2])
```

```python
col1, col2, col3, col4 = [data1[:,i] for i in range(4)]

fig1 = plt.figure(figsize=(20,10))
ax1 = fig1.add_subplot(111)
ax1.set_title("Atom Separation vs LJ Potential\nEpsilon=0.238, Sigma=3.4")    
ax1.set_xlabel('Atom Separation (Angstrom)')
ax1.set_ylabel('Potential Energy (kcal/mol)')
ax1.plot(col1, col2, label='Attractive Potential')
ax1.plot(col1, col3, label='Repulsive Potential')
ax1.plot(col1, col4, label='Overall Energy')
leg = ax1.legend()

plt.show()

fig1.savefig('./graph1.png')
```

```python
col5, col6, col7, col8 = [data2[:,i] for i in range(4)]

fig2 = plt.figure(figsize=(20,10))
ax2 = fig2.add_subplot(111)
ax2.set_title("Total LJ energy for single atom pair and 100 atom pairs\nEpsilon=0.238, Sigma=3.4")    
ax2.set_xlabel('Atom Separation (Angstrom)')
ax2.set_ylabel('Potential Energy (kcal/mol)')
ax2.plot(col1, col4, label='Total LJ Energy (1 atom pair)')
ax2.plot(col5, col8, label='Total System Energy (100 atom pairs)')
leg = ax2.legend()

plt.show()

fig1.savefig('./graph2.png')
```
