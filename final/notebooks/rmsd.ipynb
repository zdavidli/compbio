```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
wt_data = []

with open('../WT/wt_rmsd.dat') as f:
    next(f)
    for line in f:
        line = line.strip().split()
        wt_data.append([float(el) for el in line])

wt_data = np.array(wt_data[:-1])
```

```python
d102a_data = []

with open('../D102A/d102a_rmsd.dat') as f:
    next(f)
    for line in f:
        line = line.split()
        line = [float(el) for el in line]
        d102a_data.append(line)

d102a_data = np.array(d102a_data[:-1])
```

```python
plt.plot(wt_data[:, 0]/1000, wt_data[:, 1], linewidth=1, label='WT')
plt.plot(d102a_data[:, 0]/1000, d102a_data[:, 1], linewidth=1, label='D102A')
plt.xlabel("Time (ns)")
plt.ylabel(r"RMSD ($\AA$)")
plt.title('Time-Dependent RMSD of WT vs. D102A Chymotrypsin\n(Relative to Starting Structure)')
plt.yticks(np.arange(1, 3.1, 0.5))
plt.legend()
plt.savefig('../figures/rmsds.png')
```
