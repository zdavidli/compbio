```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
wt_data = []

with open('../WT/wt_rmsd.dat') as f:
    next(f)
    for line in f:
        line = line.strip().split()
        wt_data.append([float(el) for el in line])

wt_data = np.array(wt_data[:-1])
```

```python
d102a_data = []

with open('../D102A/d102a_rmsd.dat') as f:
    next(f)
    for line in f:
        line = line.split()
        line = [float(el) for el in line]
        d102a_data.append(line)

d102a_data = np.array(d102a_data[:-1])
```

```python
plt.figure(figsize=(12, 8))

plt.plot(wt_data[:, 0]/500, wt_data[:, 1], linewidth=1, label='WT')
plt.plot(d102a_data[:, 0]/500, d102a_data[:, 1], linewidth=1, label='D102A')
plt.xlabel("Time (ns)", fontsize=18)
plt.ylabel(r"RMSD ($\AA$)", fontsize=18)
plt.title('Time-Dependent RMSD of WT vs. D102A Chymotrypsin\n(Relative to Starting Structure)', fontsize=24)
plt.xticks(np.arange(0, 11), fontsize=14)
plt.yticks(np.arange(1, 3.1, 0.5), fontsize=14)
lgnd = plt.legend(prop={'size': 14})
lgnd.legendHandles[0].set_linewidth(5.0)
lgnd.legendHandles[1].set_linewidth(5.0)

plt.savefig('../figures/rmsds.eps', format='eps', dpi=1000, bbox_inches='tight')
```
