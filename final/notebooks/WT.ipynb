```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rc

import numpy as np
```

```python
data = []
with open("../WT/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        data.append([float(el) for el in line])

data = np.array(data)
```

```python
plt.figure()
plt.scatter(data[:, 0], data[:, 1], s=0.1)
plt.xlabel(r'$\chi_1^\circ$')
plt.ylabel(r'$\chi_2^\circ$')
```
