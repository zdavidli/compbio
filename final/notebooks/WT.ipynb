```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rc

import numpy as np
```

```python
data = []
with open("../WT/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        data.append([float(el) for el in line])

data = np.array(data)
```

```python
plt.figure()
plt.scatter(data[:, 0], data[:, 1], s=0.1)
plt.xlabel(r'$\chi_1^\circ$')
plt.ylabel(r'$\chi_2^\circ$')
plt.xlim((0,360))
plt.ylim((0,360))

plt.savefig('../WT/wt_asp_his.png')
```

```python
hbond = []
with open("../WT/hbonds_asp_his.dat") as f:
    for line in f:
        line = line.strip().split()
        hbond.append([float(el) for el in line])

hbond = np.array(hbond)
```

```python
fig, ax = plt.subplots()
x,y = hbond[:,0], hbond[:,1]
plt.plot(x, y, linewidth=0.25)
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
# print(fit)
plt.title("WT ASP-HIS\n"+r"$y=1.44-(2.65\times10^{-5})x$")
plt.xticks(np.arange(min(x), max(x)+2, 1000))
plt.yticks(np.arange(min(y), max(y)+1,1))

plt.savefig('../WT/wt_hbonds_asp_his.png')
```
