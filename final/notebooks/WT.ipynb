```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rc

import numpy as np
```

```python
data = []
with open("../WT/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        data.append([float(el) for el in line])

data = np.array(data)
```

```python
plt.figure()
plt.scatter(data[:, 0], data[:, 1], s=0.1)
plt.xlabel(r'$\chi_1^\circ$')
plt.ylabel(r'$\chi_2^\circ$')
plt.xlim((0,360))
plt.ylim((0,360))

# plt.savefig('../figures/wt_asp_his.png')
```

```python
hbond = []
with open("../WT/hbonds_asp_his.dat") as f:
    for line in f:
        line = line.strip().split()
        hbond.append([float(el) for el in line])

hbond = np.array(hbond)

unique, counts = np.unique(hbond[:, 1], return_counts=True)
print(dict(zip(unique, counts)))
print(4980/5000.)

plt.figure(figsize=(12, 8))
x,y = hbond[:,0]/500, hbond[:,1]
plt.plot(x, y, linewidth=0.25)
# ax.plot(x, fit[0] * x + fit[1], color='red')
# print(fit)
plt.title("WT Asp-His Hydrogen Bond Incidence", fontsize=24)#\n"+r"$y=1.44-(2.65\times10^{-5})x$")
plt.xlabel('Time (ns)', fontsize=18)
plt.ylabel("Number of hydrogen bonds", fontsize=18)
plt.xticks(np.arange(min(x), max(x)+1), fontsize=14)
plt.yticks(np.arange(min(y), max(y)+1), fontsize=14)

# plt.savefig('../figures/wt_hbonds_asp_his.eps', format='eps', dpi=1000, bbox_inches='tight')
```

```python
hbond = []
with open("../WT/hbonds_his_ser.dat") as f:
    for line in f:
        line = line.strip().split()
        hbond.append([float(el) for el in line])

hbond = np.array(hbond)

print(np.sum(hbond[:,1])/hbond.shape[0])


plt.figure(figsize=(12, 8))
x,y = hbond[:,0]/500, hbond[:,1]
plt.plot(x, y, linewidth=0.25)
fit = np.polyfit(x, y, deg=1)
# ax.plot(x, fit[0] * x + fit[1], color='red')
print(fit)
plt.title("WT His-Ser Hydrogen Bond Incidence", fontsize=24)#\n"+r"$y=0.20 + (7\times 10^{-4})x$")
plt.xlabel('Time (ns)', fontsize=18)
plt.ylabel("Number of hydrogen bonds", fontsize=18)
plt.xticks(np.arange(min(x), max(x)+1), fontsize=14)
plt.yticks(np.arange(min(y), max(y)+2), fontsize=14)
# plt.savefig('../figures/wt_hbonds_his_ser.eps', format='eps', dpi=1000,bbox_inches='tight')
```
