```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rc

import numpy as np
```

```python
d102a_data = []
with open("../D102A/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        d102a_data.append([float(el) for el in line])

d102a_data = np.array(d102a_data)
```

```python
wt_data = []
with open("../WT/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        wt_data.append([float(el) for el in line])

wt_data = np.array(wt_data)
```

```python
plt.figure()
plt.scatter(d102a_data[:, 0], d102a_data[:, 1], s=1, label='D102A')
plt.scatter(wt_data[:, 0], wt_data[:, 1], s=1, label='WT')
plt.title(r'$\chi_1$ vs $\chi_2$ in Wild-Type and D102A simulations')
plt.xlabel(r'$\chi_1^\circ$ (degrees)')
plt.ylabel(r'$\chi_2^\circ$ (degrees)')
plt.xticks(np.arange(0, 361, 90))
plt.yticks(np.arange(0, 361, 90))
plt.legend()

plt.savefig('../figures/chi_plot_wt_d102a.png')
```

```python
hbond = []
with open("../D102A/hbonds_his_ser.dat") as f:
    for line in f:
        line = line.strip().split()
        hbond.append([float(el) for el in line])

hbond = np.array(hbond)
```

```python
fig, ax = plt.subplots()
x,y = hbond[:,0]/1000, hbond[:,1]
plt.plot(x, y, linewidth=0.25)
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
print(fit)
plt.title("D102A His-Ser Hydrogen Bond Incidence\n"+r"$y=-(0.114) + (1.01\times 10^{-4})x$")
plt.xlabel('Time (ns)')
plt.ylabel("Number of hydrogen bonds")
plt.xticks(np.arange(min(x), max(x)+1))
plt.yticks(np.arange(min(y), max(y)+2,1))

plt.savefig('../figures/d102a_hbonds_his_ser.png')
```
