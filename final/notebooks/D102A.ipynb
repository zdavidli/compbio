```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import rc

import numpy as np
```

```python
d102a_data = []
with open("../D102A/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        d102a_data.append([float(el) for el in line])

d102a_data = np.array(d102a_data)
```

```python
wt_data = []
with open("../WT/ang.dat") as f:
    for line in f:
        line = line.strip().split()
        wt_data.append([float(el) for el in line])

wt_data = np.array(wt_data)
print(wt_data.shape)
```

```python
plt.figure()
plt.scatter(d102a_data[:, 0], d102a_data[:, 1], s=0.1)
plt.xlabel(r'$\chi_1^\circ$')
plt.ylabel(r'$\chi_2^\circ$')
plt.xlim((0,360))
plt.ylim((0,360))

# plt.savefig('../figures/d102a_his_angles.png')
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(d102a_data[:, 0], d102a_data[:, 1], s=2, label='D102A')
plt.scatter(wt_data[:, 0], wt_data[:, 1], s=2, label='WT')
plt.title(r'$\chi_1$ vs $\chi_2$ in Wild-Type and D102A simulations', fontsize=24)
plt.xlabel(r'$\chi_1^\circ$ (degrees)', fontsize=18)
plt.ylabel(r'$\chi_2^\circ$ (degrees)', fontsize=18)
plt.xticks(np.arange(0, 361, 90), fontsize=14)
plt.yticks(np.arange(0, 361, 90), fontsize=14)
lgnd = plt.legend(prop={'size': 14})
lgnd.legendHandles[0]._sizes = [100]
lgnd.legendHandles[1]._sizes = [100]

# plt.savefig('../figures/chi_plot_wt_d102a.eps', format='eps', dpi=1000, bbox_inches='tight')
```

```python
hbond = []
with open("../D102A/hbonds_his_ser.dat") as f:
    for line in f:
        line = line.strip().split()
        hbond.append([float(el) for el in line])

hbond = np.array(hbond)

print(np.sum(hbond[:,1])/(hbond.shape[0]))
```

```python
plt.figure(figsize=(12, 8))
x,y = hbond[:,0]/500, hbond[:,1]
plt.plot(x, y, linewidth=0.25)
# ax.plot(x, fit[0] * x + fit[1], color='red')
# print(fit)
plt.title("D102A His-Ser Hydrogen Bond Incidence", fontsize=24)#\n"+r"$y=-(0.114) + (1.01\times 10^{-4})x$")
plt.xlabel('Time (ns)', fontsize=18)
plt.ylabel("Number of hydrogen bonds", fontsize=18)
plt.xticks(np.arange(min(x), max(x)+1), fontsize=14)
plt.yticks(np.arange(min(y), max(y)+2,1), fontsize=14)

# plt.savefig('../figures/d102a_hbonds_his_ser.eps', format='eps', dpi=1000, bbox_inches='tight')
```
